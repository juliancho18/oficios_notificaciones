Segmentos de elaborar oficios


```python
# ============================================================
# 02 - Filtrar SOLO los registros con "Elaborar Oficio"
#      (columna correcta: "Estado de la Actuacion")
# ============================================================

import unicodedata

def _norm_txt(x: str) -> str:
    x = "" if x is None else str(x)
    x = x.strip().upper()
    x = " ".join(x.split())
    x = unicodedata.normalize("NFKD", x).encode("ascii", "ignore").decode("ascii")
    return x

col_estado_act = "Estado de la Actuacion"
valor_objetivo = "Elaborar Oficio"


if col_estado_act not in df.columns:
    # Si por algún motivo viene con otra escritura, mostramos candidatas
    candidatas = [c for c in df.columns if "estado" in _norm_txt(c) and "actu" in _norm_txt(c)]
    raise KeyError(
        f"No existe la columna '{col_estado_act}'.\n"
        f"Candidatas encontradas: {candidatas}\n"
        f"Columnas disponibles: {df.columns.tolist()}"
    )

tmp = df.copy()
tmp["_estado_norm"] = tmp[col_estado_act].map(_norm_txt)

df_filtrado = tmp[tmp["_estado_norm"] == _norm_txt(valor_objetivo)].drop(columns=["_estado_norm"]).copy()
df_filtrado.reset_index(drop=True, inplace=True)

print("✅ Filtrado aplicado por Estado de la Actuacion == 'Elaborar Oficio'")
print("Shape original:", df.shape)
print("Shape filtrado :", df_filtrado.shape)

display(df_filtrado.head(10))
display(df[col_estado_act].astype(str).value_counts(dropna=False).head(20))
```
