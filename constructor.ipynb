```python
def buscar_col(df, posibles):
    def norm(s):
        return str(s).strip().lower().replace(" ", "").replace("_", "").replace(".", "")
    cols = list(df.columns)
    cols_norm = {norm(c): c for c in cols}
    for p in posibles:
        if norm(p) in cols_norm:
            return cols_norm[norm(p)]
    for p in posibles:
        for c in cols:
            if norm(p) in norm(c):
                return c
    return None

col_notificador = buscar_col(df_filtrado, ["Notificador/Revisor", "Notificador"])
col_estado      = buscar_col(df_filtrado, ["Estado_Actuacion", "Estado Actuacion", "Estado Actuación", "Estado"])
col_tipo_oficio = buscar_col(df_filtrado, ["Tipo_Oficio", "Tipo de Acto Administrativo", "Tipo Oficio"])
col_expediente  = buscar_col(df_filtrado, ["numero_expediente", "No. Exp.", "Expediente"])
col_notificado  = buscar_col(df_filtrado, ["Nombre_Notificado", "Nombre Notificado", "Notificado"])

print("Detectadas:")
print("col_notificador:", col_notificador)
print("col_estado:", col_estado)
print("col_tipo_oficio:", col_tipo_oficio)
print("col_expediente:", col_expediente)
print("col_notificado:", col_notificado)
```

```python
def generar_words_desde_plantilla(
    self,
    df,
    plantilla_path,
    carpeta_salida="output_abogados",
    col_notificador="Notificador/Revisor",
    col_subcarpeta=None,                 # ej: "ACTUACION"
    nombre_archivo_cols=None,            # ej: ["numero_expediente", "Nombre Notificado"]
    col_map=None,                        # {"Numero_Documento": "Numero Documento", ...}
    abrir_cierre="«»",                   # delimitadores de tu plantilla
    max_nombre=140
):
    import os
    import re
    import pandas as pd

    try:
        from docx import Document
    except Exception as e:
        raise ImportError(
            "No se pudo importar python-docx. Instala con: pip install python-docx"
        ) from e

    if df is None or len(df) == 0:
        return pd.DataFrame(columns=["fila", "ruta_docx"])

    if not os.path.exists(plantilla_path):
        raise FileNotFoundError(f"No existe la plantilla en: {plantilla_path}")

    col_map = col_map or {}
    nombre_archivo_cols = nombre_archivo_cols or []

    def _safe_txt(x):
        if x is None:
            return ""
        if isinstance(x, float) and pd.isna(x):
            return ""
        if pd.isna(x):
            return ""
        return str(x)

    def _sanitize(s):
        s = _safe_txt(s).strip()
        s = re.sub(r"[\\/:*?\"<>|]+", "_", s)
        s = re.sub(r"\s+", " ", s).strip()
        if len(s) > max_nombre:
            s = s[:max_nombre].strip()
        return s or "SIN_DATO"

    # Reemplazo simple (si el token está partido entre runs, puede perder formato).
    # Funciona bien si tus «campos» están escritos como texto normal.
    def _replace_in_paragraph(paragraph, replacements):
        text = paragraph.text
        changed = False
        for k, v in replacements.items():
            if k in text:
                text = text.replace(k, v)
                changed = True
        if changed:
            paragraph.text = text

    def _replace_in_table(table, replacements):
        for row in table.rows:
            for cell in row.cells:
                for p in cell.paragraphs:
                    _replace_in_paragraph(p, replacements)
                for t in cell.tables:
                    _replace_in_table(t, replacements)

    izq, der = abrir_cierre[0], abrir_cierre[-1]  # « »
    generados = []

    for i, row in df.reset_index(drop=True).iterrows():

        notificador = _sanitize(row.get(col_notificador, "SIN_NOTIFICADOR"))
        subcarp = _sanitize(row.get(col_subcarpeta, "")) if col_subcarpeta else ""

        # Carpeta final:
        # carpeta_salida/<NOTIFICADOR>/<SUBCARPETA>/WORDS/
        ruta_dir = os.path.join(carpeta_salida, notificador)
        if subcarp:
            ruta_dir = os.path.join(ruta_dir, subcarp)
        ruta_dir = os.path.join(ruta_dir, "WORDS")
        os.makedirs(ruta_dir, exist_ok=True)

        # Nombre del archivo
        partes = []
        for c in nombre_archivo_cols:
            partes.append(_sanitize(row.get(c, "")))
        base_name = "_".join([p for p in partes if p and p != "SIN_DATO"]) or f"doc_{i+1}"
        nombre_doc = f"{base_name}.docx"
        ruta_out = os.path.join(ruta_dir, nombre_doc)

        # Construir replacements desde la fila
        replacements = {}

        # 1) Todos los campos directos que existan en el DF:
        for col in df.columns:
            key = f"{izq}{col}{der}"     # «Nombre Notificado»
            replacements[key] = _safe_txt(row.get(col, ""))

        # 2) Map de plantilla->col real (para casos como Numero_Documento vs Numero Documento)
        for campo_plantilla, col_real in col_map.items():
            key = f"{izq}{campo_plantilla}{der}"  # «Numero_Documento»
            replacements[key] = _safe_txt(row.get(col_real, ""))

        # Cargar plantilla y reemplazar
        doc = Document(plantilla_path)

        for p in doc.paragraphs:
            _replace_in_paragraph(p, replacements)

        for t in doc.tables:
            _replace_in_table(t, replacements)

        doc.save(ruta_out)
        generados.append({"fila": i, "ruta_docx": ruta_out})

    return pd.DataFrame(generados)
```

```python
from docx import Document
from pathlib import Path
import pandas as pd

ruta_plantillas = Path("plantillas")

def extraer_texto_docx(ruta_docx):
    doc = Document(str(ruta_docx))
    partes = []

    for p in doc.paragraphs:
        if p.text.strip():
            partes.append(p.text.strip())

    for t in doc.tables:
        for row in t.rows:
            for cell in row.cells:
                for p in cell.paragraphs:
                    if p.text.strip():
                        partes.append(p.text.strip())

    return "\n".join(partes)

registros = []

for archivo in ruta_plantillas.glob("*.docx"):
    texto = extraer_texto_docx(archivo)

    tiene_marcadores = any(x in texto for x in ["«", ">>", "{{", "["])
    tiene_datos_quemados = any(x in texto for x in [
        "40381",
        "41480",
        "01237000397",
        "50237000469",
        "giovannipadillatellez@gmail.com",
        "comunicacionespgn@procuraduria.gov.co",
        "YUBER YESID CARDENAS PULIDO",
        "PROCURADORIA JUDICIAL"
    ])

    registros.append({
        "archivo": archivo.name,
        "tiene_marcadores": tiene_marcadores,
        "tiene_datos_quemados": tiene_datos_quemados,
        "muestra_texto": texto[:300]
    })

df_revision_plantillas = pd.DataFrame(registros)
display(df_revision_plantillas)
```

```python
import shutil
from pathlib import Path
from utils import Utils

utils = Utils()

# Limpiar salida final
ruta_final = Path("output_abogados")
if ruta_final.exists():
    shutil.rmtree(ruta_final)

# Base final solo para oficios válidos
COL_ESTADO = "Estado de la Actuacion"
COL_ACTUACION = "ACTUACION"

df_oficios = df_filtrado.copy()

df_oficios = df_oficios[
    df_oficios[COL_ESTADO].astype(str).str.strip().str.lower().eq("elaborar oficio")
].copy()

no_validos = ["EJECUTORIA", "EDICTO", "EN CONSULTA DR"]
mask_no_validos = df_oficios[COL_ACTUACION].astype(str).str.upper().apply(
    lambda x: any(p in x for p in no_validos)
)
df_oficios = df_oficios[~mask_no_validos].copy()

df_oficios = utils.agregar_info_regional(df_oficios, col_regional="Regional")

df_oficios["Numero_Documento"] = df_oficios["Numero Documento"]
df_oficios["Correo_Electronico"] = df_oficios["Correo Electronico"]
df_oficios["No_DireccionEfectiva"] = df_oficios["No. DireccionEfectiva"]
df_oficios["Tipo_de_documento"] = df_oficios["Tipo de documento"]
df_oficios["Funcionario_que_expide"] = df_oficios["Funcionario que expide"]
df_oficios["Nombre_Notificado"] = df_oficios["Nombre Notificado"]

def construir_modo_envio(row):
    actuacion = str(row.get("ACTUACION", "")).strip().upper()
    autoriza = str(row.get("Autoriza Notificacion electrónica", "")).strip().lower()
    correo = str(row.get("Correo Electronico", "")).strip()

    if "NOTIFICACIÓN ELECTRÓNICA" in actuacion or "NOTIFICACION ELECTRONICA" in actuacion:
        return "ELECTRÓNICO"
    if autoriza in ["si", "sí", "s", "true", "1", "x"] and correo != "":
        return "ELECTRÓNICO"
    return "FÍSICO"

df_oficios["Modo_de_envío"] = df_oficios.apply(construir_modo_envio, axis=1)
df_oficios["Modo_de_envio"] = df_oficios["Modo_de_envío"]

def homologar_actuacion(x):
    txt = str(x).strip().upper()
    if "AVISO" in txt:
        return "AVISO"
    if "CITACION" in txt:
        return "CITACION"
    if "COMUNICACION" in txt:
        return "COMUNICACION"
    if "NOTIFICACIÓN ELECTRÓNICA" in txt or "NOTIFICACION ELECTRONICA" in txt:
        return "NOTIFICACIÓN ELECTRÓNICA"
    return txt

df_oficios["ACTUACION_NORMALIZADA"] = df_oficios["ACTUACION"].apply(homologar_actuacion)

resumen_final = utils.generar_words_desde_tipo_oficio(
    df=df_oficios,
    carpeta_salida="output_abogados",
    col_notificador="Notificador/Revisor",
    col_estado_actuacion="Estado de la Actuacion",
    col_tipo_oficio="ACTUACION_NORMALIZADA",
    col_subcarpeta=None,
    nombre_archivo_cols=["numero_expediente", "Nombre_Notificado"],
    col_map={}
)

display(resumen_final.head(20))
print("Total generados:", len(resumen_final))

if "error" in resumen_final.columns:
    errores = resumen_final[resumen_final["error"].notna()]
    print("Total errores:", len(errores))
    if len(errores) > 0:
        display(errores.head(50))
```

```python
# =========================
# RESUMEN TIPO TABLA DINÁMICA
# =========================

df_resumen_ok = resumen_final.copy()

# Nos quedamos solo con registros generados correctamente
if "error" in df_resumen_ok.columns:
    df_resumen_ok = df_resumen_ok[df_resumen_ok["error"].isna()].copy()

# Estandarizar nombres por seguridad
if "abogado" in df_resumen_ok.columns:
    df_resumen_ok["abogado"] = df_resumen_ok["abogado"].astype(str).str.strip()

if "tipo_oficio" in df_resumen_ok.columns:
    df_resumen_ok["tipo_oficio"] = df_resumen_ok["tipo_oficio"].astype(str).str.strip().str.upper()

# 1) Tabla detalle: quién creó qué tipo y cuántos
tabla_detalle = (
    df_resumen_ok
    .groupby(["abogado", "tipo_oficio"], dropna=False)
    .size()
    .reset_index(name="cantidad_oficios")
    .sort_values(["abogado", "tipo_oficio"])
)

print("RESUMEN DETALLADO POR PERSONA Y TIPO DE OFICIO")
display(tabla_detalle)

# 2) Tabla dinámica: filas = abogado, columnas = tipo_oficio
tabla_dinamica = pd.pivot_table(
    df_resumen_ok,
    index="abogado",
    columns="tipo_oficio",
    values="archivo",
    aggfunc="count",
    fill_value=0
)

# Agregar total por fila
tabla_dinamica["TOTAL"] = tabla_dinamica.sum(axis=1)

# Ordenar por total descendente
tabla_dinamica = tabla_dinamica.sort_values("TOTAL", ascending=False)

print("TABLA DINÁMICA DE OFICIOS GENERADOS")
display(tabla_dinamica)

# 3) Resumen general por tipo de oficio
resumen_por_tipo = (
    df_resumen_ok["tipo_oficio"]
    .value_counts(dropna=False)
    .reset_index()
)
resumen_por_tipo.columns = ["tipo_oficio", "cantidad_total"]

print("RESUMEN GENERAL POR TIPO DE OFICIO")
display(resumen_por_tipo)

# 4) Resumen general por persona
resumen_por_persona = (
    df_resumen_ok["abogado"]
    .value_counts(dropna=False)
    .reset_index()
)
resumen_por_persona.columns = ["abogado", "cantidad_total"]

print("RESUMEN GENERAL POR PERSONA")
display(resumen_por_persona)
```
